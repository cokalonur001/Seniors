```python
class ToDoList:
    def __init__(self, tasks):
        self.tasks = tasks

    def get_most_efficient_time(self):
        completed_tasks = []
        incomplete_tasks = []
        for task in self.tasks:
            if task["completed"]:
                completed_tasks.append(task)
            else:
                incomplete_tasks.append(task)

        if len(completed_tasks) == 0:
            return "No completed tasks found."

        total_time = 0
        for task in completed_tasks:
            total_time += task["time"]

        avg_time_per_task = total_time / len(completed_tasks)

        min_time = float("inf")
        min_time_task = None
        for task in incomplete_tasks:
            if task["time"] < min_time:
                min_time = task["time"]
                min_time_task = task

        if min_time_task is None:
            return "No incomplete tasks found."

        return min_time_task["name"], min_time, avg_time_per_task


def main():
    tasks = []
    while True:
        task_name = input("Enter task name (or 'done' to finish): ")
        if task_name == "done":
            break

        try:
            task_time = int(input("Enter task time in minutes: "))
        except ValueError:
            print("Invalid input. Please enter a valid integer.")
            continue

        if task_time < 0:
            print("Invalid input. Please enter a positive integer.")
            continue

        task_completed = False
        while True:
            task_status = input("Is the task completed (y/n)? ")
            if task_status == "y":
                task_completed = True
                break
            elif task_status == "n":
                break
            else:
                print("Invalid input. Please enter 'y' or 'n'.")

        task = {"name": task_name, "time": task_time, "completed": task_completed}
        tasks.append(task)

    todo_list = ToDoList(tasks)

    try:
        max_time = int(input("What is the maximum amount of time you have available (in minutes)? "))
    except ValueError:
        print("Invalid input. Please enter a valid integer.")
        return

    if max_time < 0:
        print("Invalid input. Please enter a positive integer.")
        return

    efficient_task = todo_list.get_most_efficient_time()

    if efficient_task[1] <= max_time:
        print(efficient_task[0] + " is the most efficient task to complete next. It will take " + str(efficient_task[1]) + " minutes, while the average time per completed task is " + str(efficient_task[2]) + " minutes.")
        with open("completed_tasks.csv", "a") as f:
            f.write(efficient_task[0] + ": completed\n")
    else:
        print("You do not have enough time to complete any task.")
        with open("completed_tasks.csv", "a") as f:
            f.write("No tasks completed\n")

    with open("completed_tasks.csv", "r") as f:
        contents = f.read()
        print(contents)

if __name__ == "__main__":
    main()
```
